## Cache Plots

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
# from matplotlib.pyplot import figure
import pandas as pd
import seaborn as sns

def plot(p_path, p_yId):
    data_frame = pd.read_csv(p_path)

    # data_frame = data_frame[data_frame.n_tiling_cols >= 4]
    data_frame = data_frame[data_frame.n_tiling_cols < 65536]
    # data_frame = data_frame[data_frame.n_tiling_levels_rows <= 16]
    data_frame['n_tiling_levels_rows'] = data_frame['n_tiling_levels_rows'].map(str)
    
    rcParams['figure.figsize'] = 11.7,8.27
    plot = sns.lineplot(x='n_tiling_cols',
                        y=p_yId,
                        hue='n_tiling_levels_rows',
                        style='impl_id',
                        data=data_frame)
    plot.set_yscale('log')
    plot.set_xscale('log')
    plot.set(xlabel='n_tiling_cols', ylabel=p_yId)
    plt.figure(figsize=(1, 1), dpi=80)
    plt.show()
```

### Gauss3

#### Cube 2050

```python
path_cube_gauss3 = './exe_op_bench_linear_stencil_likwid_cache_cube2050_gauss3.csv'
plot(path_cube_gauss3,'actual_cpu_clock')
plot(path_cube_gauss3,'max_cpu_clock')
plot(path_cube_gauss3,'retired_instructions')
plot(path_cube_gauss3,'cpu_clocks_unhalted')
plot(path_cube_gauss3,'data_cache_accesses')
plot(path_cube_gauss3,'data_cache_refills_all')
plot(path_cube_gauss3,'runtime')
# plot(path_cube_gauss3,'runtime_unhalted')
plot(path_cube_gauss3,'clock')
plot(path_cube_gauss3,'cpi')
plot(path_cube_gauss3,'data_cache_requests')
plot(path_cube_gauss3,'data_cache_request_rate')
plot(path_cube_gauss3,'data_cache_misses')
plot(path_cube_gauss3,'data_cache_miss_rate')
plot(path_cube_gauss3,'data_cache_miss_ratio')
```
