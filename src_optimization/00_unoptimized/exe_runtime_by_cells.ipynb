## Runtime by cells

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
# from matplotlib.pyplot import figure
import pandas as pd
import seaborn as sns

def plotRoutine(p_path):
    data_frame = pd.read_csv(p_path)
    data_frame = data_frame[data_frame.factor < 1000]
    # data_frame = data_frame[data_frame.node_id == "node2"]
    
    rcParams['figure.figsize'] = 11.7,8.27
    plot = sns.lineplot(x='n_cells',
                        y='runtime',
                        hue='impl_id',
                        style='time_id',
                        data=data_frame)
    plot.set_yscale('log')
    plot.set_xscale('log')
    plot.set(xlabel='Number of cells', ylabel='Runtime [s]')
    plt.figure(figsize=(1, 1), dpi=80)
    plt.show()
```

### Gauss3

```python
path = './exe_runtime_by_cells_gauss3.csv'
plotRoutine(path)
```
